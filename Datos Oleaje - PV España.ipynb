**En el presente cuaderno se describe el dataset "datos de oleaje" a través de estadísticos y gráficos (linea, boxplot y scatterplot).**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dir_archivo = "C:/Users/nubem/Documents/OLEAJE/dataset.xlsx" #Ruta del archivo

df_per_vb = pd.read_excel(dir_archivo, sheet_name='PER-VB') #Periodo Boya de Bilbao-Vizcaya 
df_per_b2 = pd.read_excel(dir_archivo, sheet_name='PER-B2') #Periodo Boya Costera de Bilbao 
df_per_ab = pd.read_excel(dir_archivo, sheet_name='PER-AB') #Periodo Boya de Abra-Ciervana
df_hsmax_vb = pd.read_excel(dir_archivo, sheet_name='HSMAX-VB') #Altura significativa de oleaje Boya de Bilbao-Vizcaya  
df_hsmax_b2 = pd.read_excel(dir_archivo, sheet_name='HSMAX-B2') #Altura significativa de oleaje Boya Costera de Bilbao 
df_hsmax_ab = pd.read_excel(dir_archivo, sheet_name='HSMAX-AB') #Altura significativa de oleaje Boya de Abra-Ciervana
df_hsdir_vb = pd.read_excel(dir_archivo, sheet_name='HSDIR-VB') #Dirección del oleaje? Boya de Bilbao-Vizcaya  
df_hsdir_b2 = pd.read_excel(dir_archivo, sheet_name='HSDIR-B2') #Dirección del oleaje? Boya Costera de Bilbao 
df_hsdir_ab = pd.read_excel(dir_archivo, sheet_name='HSDIR-AB') #Dirección del oleaje? Abra-Ciervana
df_comax_vb = pd.read_excel(dir_archivo, sheet_name='COMAX-VB') #Corriente máxima Bilbao-Vizcaya 
df_codir_vb = pd.read_excel(dir_archivo, sheet_name='CODIR-VB') #Dirección corriente Bilbao-Vizcaya 
df_nmax = pd.read_excel(dir_archivo, sheet_name='NMAX') #Nivel máximo Bilbao
df_nmin = pd.read_excel(dir_archivo, sheet_name='NMIN') #Nivel mínimo Bilbao
df_nmed = pd.read_excel(dir_archivo, sheet_name='NMEDIO') #Nivel medio Bilbao
```
